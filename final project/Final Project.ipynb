```python
import pandas as pd
import numpy as np
import geopandas
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("medianageofdeath1.csv")
neighborhoods = geopandas.read_file("Neighborhoods/Neighborhoods_.shp")
ages = data.groupby("NEIGHBORHOOD").sum()['TOTAL MD AGE AT DEATH']

agesMap = neighborhoods.merge(ages, left_on='hood', right_on='NEIGHBORHOOD')
agesMap.plot(column='TOTAL MD AGE AT DEATH',
               edgecolor="white",
               legend=True,
               legend_kwds={'label': "Median Age at Death"},
               figsize=(15, 10),
               missing_kwds={"color": "lightgrey"})
```

The top ten highest median ages at death are as follows (ignoring the one where it says 0):

```python
agesSorted = ages.sort_values(ascending=False)
agesSorted.head(10)
```

This points to neighborhoods like North Oakland and Squirrel Hill South being the best. High life expectancy can point to many things, among which are access to quality healthcare, personal safety (lower crime rates), and lower rates of pollution.
